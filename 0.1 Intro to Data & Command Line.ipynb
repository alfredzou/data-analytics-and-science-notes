### Data Analytics
---
* Extracting insights from data to inform decision making  
* Data analytics is the blend of 4 major components: 
    1. Understanding the subject matter or business to ask the right question. We need to come up with hypotheses, which we will test by looking into the data. 
    2. Programming skills to clean and analyse data. While we prepare the data we need to ask ourselves the questions like 'will I be able to explore the hypothesis using this data?' and 'is this data trustworthy?'. During this process we may come up with new hypotheses through exploratory analysis of the data
    3. Statistical know how to have the confidence the insights you found are stastically sound
    4. Ability to communicate insights & provide recommendations to stakeholders

##### Absolute Path vs Relative Path for Directory Navigation
* The root is the lowest level of all folders, with all other folders branching off the root path. It is denoted by `/`
* The home directory is a special folder denoted by `~`. This is usually found at `/Users/[Your Username]`
* When navigating directories, we use `cd` with either an absolute or relative path.
    * Absolute paths start from the root
    * Relative paths start from the current working directory
        * `pwd`: prints the current working directory
        * `ls`: list files in current directory
        * `ls *w*` - searches for all files including the letter w in the letter. This is an example of using wild cards
* `cd /`: navigate to root
* `cd`: navigate to home directory
* `cd folder_name`: navigate to folder (relative path)
* `cd /..`: navigate to folder (absolute path)
* `cd -`: navigate to preivous directory
* `cd + Tab`: use auto complete for difficult folder names
* `cd "General Assembly`": enclose folder names with special characters
* `cd General\ Assembly`: use a back slash to escape special characters
* `cd + drag and drop folder`: dropping a folder is an alternative method

##### File Manipulation
* Commands for creating, deleting, copying, moving and renaming
* `start file_name`: open file
* `touch file_name`: create file
* `rm file_name`: permanently delete file
* `mkdir folder_name`: create folder
* `rmdir folder_name`: delete empty folder
* `rm -r folder_name`: delete folders and their contents recursively
* `cp [initial_file_location] [final_file_location]`: copys a file
* `mv [initial_file_location] [final_file_location]`: moves a file. Can also be used for renaming

##### Commands
* Bash can use commands. Here are a few useful commands for reading from and writing to files, etc.
* `command --help`: opens help for associated command
* `tldr command`: opens a short version of help. Must be installed first using `pip install tldr`
* `cat file_name`: prints file 
* `head file_name`: prints top 10 lines 
* `head file_name -n 3`: prints top 3 lines 
* `tail file_name`: prints bottom 10 lines
* `echo 'text' > file`: overwrites with given text
* `echo 'text' >> file`: appends to the end of the file
* `cat < file1.txt > file2.txt`: takes the input from file1.txt and writes it to file2.txt 
* `python py_file.py`: runs file in python
* `wc -l file_name`: counts number of lines in file. wc stands for word count
* `history`: will show the history of all previously run commands
* `!n`: will run the nth command from history
* `nano file_name`: opens the file

##### grep and find
* To find files use `find`. To search for text in files use `grep`
* `find`: lists every folder and file recursively 
* `find -type f`: lists every file
* `find -type d`: lists every directory
* `find -name my_name`: filters by name (case sensitive)
* `find -iname my_name`: filters by name (case insensitive)
* `find -name "my_name1" -and -name "my_name2"`: filtering by two conditions with and operator
* `find -name "my_name1" -or -name "my_name2"`: filtering by two conditions with or operator
* `find -not -name "my_name"`: filtering with not operator
* `find \( -name "*.js" -type f \) -or -type d`: grouping using `()`, which must be escaped with `\`
* `find -exec command {} \;`: executes command for each path one by one (runs one command per path)
* `find -exec command {} +`: appends each path and feeds them in as arguments for one command (runs one command). May run into issues with maximum character limit
* `grep -r string *`: searches for the string within the folder and its contents recursively
* `grep -l string *`: lists all files that contain the string
* `grep string file_name`: searches for the string within a file  
* `grep -P string file_name`: searches for the string within a file (using perl regex)
* `grep -i string file_name`: searches for the string within a file (case insensitive)
* `grep -n string file_name`: searches for the string within a file (with line numbers)

##### Commands and Paths
* When a command is run, bash searches all the user and system paths for the specified command
* If a path does not exist, the user can edit the system variables and add it to either the user or system variable path
* `echo $PATH`: prints the paths (user variable paths and system variable paths)
* `which command`: prints the path of the command that would be run 

##### Piping
* Takes an output of one command and uses it for the input of another command
* `history | grep string`: searches the history with the string
* `ls | grep string`: searches the directory with the string
* `echo $PATH | tr ":" "\n"`: returns the `PATH` variable and transforms all colons to new lines for easier reading

##### Environment Commands (conda)
* Environments are useful for package management
* Since they are separated from each other, its easy to manage packages and prevent pack conflicts with other environments
* `conda activate my_env`: activates an environment
* `conda deactivate`: deactivates the environment
* `conda env list`: list of all environments and paths
* `conda install my_package`: installs a package using conda into the active environment
* `install install my_package`: installs a package using pip into the active environment
* `conda create --name my_env`: creates an environment
* `conda remove --name my_env -all`: deletes an environment
* `pip freeze`: view all packages installed in the current environment

##### Environment Commands (venv)
* `venv` is useful for creating lightweight virtual environments
* Specifically, good for testing dependencies for scripts
* `python -m venv my_virtual_environment_path`: to create an environment at the specified path
* `source my_virtual_environment_path/Scripts/activate`: to activate the environment at the specified path. `Scripts` for windows
* `deactivate`: deactivate the `venv` environment

### Bash Scripting

##### Bash Basics
* Some programming related commands
* `foo=bar`: variable assignment. Note if we do `foo = bar`, we are passing the arguments `=` and `bar` into the command foo. Therefore, don't uses spaces during variable assignment
* `$foo` → bar: when calling variable, you need to prefix it with `$`
* `echo "$foo"` → bar : "" for string formating
* `echo '$foo'` → $foo : '' prints the literal text
* `echo foo{1,3}` → foo1 foo3
* `echo foo{1..3}` → foo1 foo2 foo3
* `?`: single character wildcard
* `*`: multiple character wildcard
* `!!`: last command
* `expr 30 + 10` → 40: add 
* `expr 30 - 10` → 20: subtract
* `expr 30 / 10` → 3: divide. Note that bash cannot handle floating points, only integers
* `expr 30 \* 10` → 300: multiply. Need to escape `*` it using `\` 
* `let my_num2=my_num1+10`: variable assignment based on another variable
* `read user_variable`: prompting for user input
* `echo $?`: check exit code
* → 0: success, otherwise its a failure (1-255)
* `$(my_variable/pattern/string)`: replace pattern with string

##### Comparison operators
* Numeric operators
* `-eq`: is equal to
* `-ne`: is not equal to
* `-gt`: greater than
* `-ge`: greater than or equal to
* `-lt`: less than 
* `-le`: less than or equal to
* String operators
* `=`: is equal to
* `!=`: is equal to
* Testing if a file or folder exists
* `-f path`: tests if a file exists
* `-d path`: tests if a folder exists

##### Bash Scripting
* Bash scripts should be prefixed with the shebang `#!/bin/bash`. This tells LINUX systems to interpret the script using the interpretor in the first line. 
* Standard input (`stdin`) - the input to your command
* Standard output (`stdout`) - anything shown in the terminal thats not an error
* Standard error (`stderr`) - errors received, if any
* `ls /root 1> output.log 2> error.log`: redirects the standard output and errors into  respective logs. `1>`, stream 1, for standard outputs and `2>`, stream 2, for standard errors
