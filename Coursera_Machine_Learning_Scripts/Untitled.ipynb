##### Classificaiton
* sigmoid/logistic function $g(z)=\frac{1}{1+e^{-z}}$
* $h_\theta(x)=g(\theta^Tx)=\frac{1}{1+e^{-\theta^Tx}}$
* $h_\theta(x)$ gives the probability of y=1
* however we need a decision boundary

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
z = np.linspace(-10,10,1000)
```

```python
g = [1/(1+np.e**(-i)) for i in z]
```

```python
plt.plot(z,g)
```

```python
z = np.linspace(-100,100,1000)
```

```python
g = [1/(1+np.e**(-i/10)) for i in z]
```

```python
plt.plot(z,g)
```

```python
h_x = np.linspace(0.001,0.999,1000)
```

```python
y_0 = [-np.log(i) for i in h_x]
```

```python
y_1 = [-np.log(1-i) for i in h_x]
```

```python
plt.plot(h_x,y_0)
```

```python
plt.plot(h_x,y_1)
```

```python
plt.plot(h_x,y_0)
plt.plot(h_x,y_1)
```

* $cost(h_\theta(X),y) = -ylog(h_\theta(x))-(1-y)log(1-h_\theta(x))$
* $J = \sum_{i=1}^{m}cost(h_\theta(X),y) = -\frac{1}{m}\sum_{i=1}^{m}[ylog(h_\theta(x))+(1-y)log(1-h_\theta(x)]$
* minimise cost to find $\theta$ 
* To make a prediction we use the $\theta$ for $h_\theta(x)=\frac{1}{1+e^{-\theta^Tx}}$

##### Regularization

* hard to prevent overfitting with lots of features
1. Reduce features
    * manually select which features to keep (you lose information but)
    * model selection alogirthm
2. Regularization
    * keep all the features but reduce their parameter values $\theta_j$
* smaller values 
* simpler hypothesis
* less prone to overfitting
* $J(\theta) = \frac{1}{2m}[\sum_{i=1}^m(h_\theta(x^{(i)}-y^{(i)})^2+\lambda\sum_{j=1}^n\theta_j^2]$
* if $\lambda$ is too large, it will be underfitted
* if $\lambda$ is too small, it will be overfitted
